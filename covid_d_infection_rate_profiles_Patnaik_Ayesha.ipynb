## **Task #2: Plot the profile rates per 1M residents**

Since the countries populations vary a lot, counting the absolute cases and deaths is not a fair metric. Normalizing the cases and deaths per 1 million residents is a typical procedure to get an unbiased view of the data.

a) Cumulative infection rate per 1M

For counting cumulative infection rate, I am taking top 25 countries i.e. N=25 and day 0 at 100 cases i.e. K=100 as is discussed in part 1 of the assignment.

Importing relevant packages

```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import matplotlib.dates as mdates
from adjustText import adjust_text
from matplotlib.ticker import ScalarFormatter
```

```python
#Loading the master cumulative table that was created in first part
covid_cumulative = pd.read_excel("/Users/AyeshaPatnaik/Downloads/UofT/Sem2_Winter-23/CHE1148/Assignment-1/Code generated files/covid_a_master_cumulative_table.xlsx")
```

```python
covid_cumulative
```

Calculating the last der for each country

```python
last_day_per_country = covid_cumulative.groupby(by=['Countries'])['Date'].max() # type(last_day_per_country) is pandas.Series
last_day_per_country = last_day_per_country.to_frame() # We convert it to pandas.DataFrame
last_day_per_country.reset_index(inplace=True) # And reset the index so that we can further join on 'Countries'
```

Finding the top 25 infected countries

```python
N = 25
last_day = pd.merge(covid_cumulative, last_day_per_country, how='inner', left_on = ['Countries', 'Date'], right_on = ['Countries', 'Date'])
last_day = last_day.sort_values('Cases', ascending=False)
last_day
top_countries = last_day[:N]['Countries'].tolist()
top_countries #top N countries as list
```

Sorting the countries with more than 100 cases

```python
K = 100
covid_top_countries = covid_cumulative[covid_cumulative.Countries.isin(top_countries)]
covid_top_countries = covid_top_countries[covid_top_countries['Cases'] > K]
```

Finding the day 0 for each country

```python
temp_day_0_for_every_country = covid_top_countries.groupby(by=['Countries']).min()
temp_day_0_for_every_country = temp_day_0_for_every_country['Date']
day_0_for_every_country=temp_day_0_for_every_country.to_frame()
```

```python
covid_top_countries = covid_top_countries.merge(day_0_for_every_country, how='inner', on='Countries')
covid_top_countries.rename(columns = {'Date_y': 'Day_0'}, inplace = True)
covid_top_countries['Days_from_0'] = (covid_top_countries['Date_x'] - covid_top_countries['Day_0']).dt.days

covid_top_countries
```

Calculating per million cases 

```python
covid_top_countries["Cases_per_1m"]=covid_top_countries["Cases"]/covid_top_countries["Population"]*1000000
covid_top_countries.drop(columns=['Unnamed: 0'],axis=1,inplace=True)
covid_top_countries.set_index('Countries', drop=False, inplace=True)
covid_top_countries.index.name = None
covid_top_countries
```

Creating the df  with the maximum cumulatve cases value per country and sorting them in descending order. This way, we then filter appropriately to create the 4 dataframes data_1_5, data_6_10, data_11_15 and data_16_20 to plot them in the 4 subplots below.

```python
max_cum_cases = covid_top_countries.groupby(by=['Countries']).max()
max_cum_cases = max_cum_cases.sort_values('Cases_per_1m', ascending=False)
max_cum_cases['Countries'] = max_cum_cases.index

data_1_5 = covid_top_countries[covid_top_countries.Countries.isin(max_cum_cases[:5]['Countries'].tolist())]
countries_1_5 = max_cum_cases[:5]['Countries'].tolist()
data_6_10 = covid_top_countries[covid_top_countries.Countries.isin(max_cum_cases[5:10]['Countries'].tolist())]
countries_6_10 = max_cum_cases[5:10]['Countries'].tolist()
data_11_15 = covid_top_countries[covid_top_countries.Countries.isin(max_cum_cases[10:15]['Countries'].tolist())]
countries_11_15 = max_cum_cases[10:15]['Countries'].tolist()
data_16_20 = covid_top_countries[covid_top_countries.Countries.isin(max_cum_cases[15:20]['Countries'].tolist())]
countries_16_20 = max_cum_cases[15:20]['Countries'].tolist()
max_cum_cases
```

```python
#Reindexing the dataframes to avoid duplicate indices during plotting

data_1_5.reset_index(inplace=True)
data_6_10.reset_index(inplace=True)
data_11_15.reset_index(inplace=True)
data_16_20.reset_index(inplace=True)
```

```python
sns.set(rc={'figure.figsize':(16,16)}, font_scale = 2)
fig, axes = plt.subplots(2, 2)
fig.tight_layout(h_pad = 2, w_pad=2)

sns.lineplot(x="Days_from_0", y="Cases_per_1m", hue="Countries", data=data_1_5, markers=True, hue_order=countries_1_5, legend=False, ax=axes[0, 0])
texts = [axes[0, 0].text(max_cum_cases.loc[str(country)]['Days_from_0'], covid_top_countries.loc[str(country)]['Cases_per_1m'].iloc[-1], str(country), fontsize=16, color=axes[0, 0].get_lines()[countries_1_5.index(country)].get_c(), ha = 'left', va = 'center') for country in countries_1_5]
adjust_text(texts)
axes[0, 0].set_ylabel("Cases per 1m-trend",fontsize=24)
axes[0, 0].set_title("Data_1_5")

sns.lineplot(x="Days_from_0", y="Cases_per_1m", hue="Countries", data=data_6_10, markers=True, hue_order=countries_6_10, legend=False, ax=axes[0, 1])
texts = [axes[0, 1].text(max_cum_cases.loc[str(country)]['Days_from_0'], covid_top_countries.loc[str(country)]['Cases_per_1m'].iloc[-1], str(country), fontsize=16, color=axes[0, 1].get_lines()[countries_6_10.index(country)].get_c(), ha = 'left', va = 'top') for country in countries_6_10]
adjust_text(texts)
axes[0, 1].set_title("Data_6_10")

sns.lineplot(x="Days_from_0", y="Cases_per_1m", hue="Countries", data=data_11_15, markers=True, hue_order=countries_11_15, legend=False, ax=axes[1, 0])
texts = [axes[1, 0].text(max_cum_cases.loc[str(country)]['Days_from_0'], covid_top_countries.loc[str(country)]['Cases_per_1m'].iloc[-1], str(country), fontsize=16, color=axes[1, 0].get_lines()[countries_11_15.index(country)].get_c(), ha = 'left', va = 'center') for country in countries_11_15]
adjust_text(texts)
axes[1, 0].set_ylabel("Cases per 1m-trend",fontsize=24)
axes[1, 0].set_xlabel("Days from reaching 100 cases",fontsize=24)
axes[1, 0].set_title("Data_11_15")

sns.lineplot(x="Days_from_0", y="Cases_per_1m", hue="Countries", data=data_16_20, markers=True, hue_order=countries_16_20, legend=False, ax=axes[1, 1])
texts = [axes[1, 1].text(max_cum_cases.loc[str(country)]['Days_from_0'], covid_top_countries.loc[str(country)]['Cases_per_1m'].iloc[-1], str(country), fontsize=16, color=axes[1, 1].get_lines()[countries_16_20.index(country)].get_c(), ha = 'left', va = 'center') for country in countries_16_20]
adjust_text(texts)
axes[1, 1].set_xlabel("Days from reaching 100 cases",fontsize=24)
axes[1, 1].set_title("Data_16_20")

axes[0, 1].set_ylabel("",fontsize=24)
axes[1, 1].set_ylabel("",fontsize=24)
axes[0, 0].set_xlabel("",fontsize=24)
axes[0, 1].set_xlabel("",fontsize=24)
axes[0, 1].set_xlim([0, 200])
axes[1, 1].set_xlim([0, 200])
axes[0, 0].set_xlim([0, 200])
axes[1, 0].set_xlim([0, 200])

plt.savefig('Fig_2_1_Patnaik_Ayesha.png',bbox_inches = 'tight')
```
